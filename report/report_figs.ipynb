```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random y values between 0.71 and 0.79
y_values = np.array([0.75, 0.80, 0.73, 0.76, 0.85, 0.79])

# Set x values
x_values = ['Imbalanced. Dense 4.\nEpoch 10 Batch 100', #0.79
            'Imbalanced. Dense 4.\nEpoch 10 Batch 250', #0.80
            'Imbalanced. Dense 7.\nEpoch 20 Batch 250', #0.71
            'Balanced. Dense 4.\nEpoch 10 Batch 100',
            'Balanced. Dense 4.\nEpoch 20 Batch 250',
            'Balanced. Dense 7.\nEpoch 10 Batch250']

# Create the line graph
plt.plot(x_values, y_values, linewidth=3)
plt.xticks(rotation=90)
plt.xlabel('Conditions')
plt.ylabel('F1 Scores')
plt.title('LSTM F1 Scores')
plt.savefig('LSTM F1 Scores.png', bbox_inches='tight')
plt.show()
```
