Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset = pd.read_csv('cleaned_and_labeled_dataset.csv')
```

Number of unique conditions and drugs

```python
condition_unique_value_count = len(dataset['condition'].unique())
print(f'Number of unique conditions: {condition_unique_value_count}')
drug_unique_value_count = len(dataset['drugName'].unique())
print(f'Number of unique drugs: {drug_unique_value_count}')
```

Distribution of the number of ratingsLabel

```python
dataset['ratingLabel'].value_counts().plot(kind='bar', ylim=0)
plt.show()
```

```python
dataset.groupby('ratingLabel').review.count().plot.bar(ylim=0, xlabel='Rating Label', ylabel='Number of Reviews',
                                                       title='Distribution of the number of ratingsLabel', rot=0)
plt.ylabel('Number of Reviews',labelpad=-1)
plt.savefig('rating_label_distribution.png', dpi=300)
plt.show()
dataset.groupby('ratingLabel').review.count()
```

Distrubiton of rating

```python
dataset.groupby('rating')['ratingLabel'].value_counts().unstack().plot(kind='bar', stacked=True, xlabel='Rating', ylabel='Number of Reviews', title='Distrubution of Ratings', rot=0)
plt.savefig('rating_distribution.png', dpi=300)
plt.show()
```

```python
grouped = dataset.groupby(['condition'])['drugName'].nunique()

# Count the number of conditions that have the same number of unique drug names
counts = grouped.value_counts()

# Plot the bar chart
counts.hist(bins=50)

# Add x and y labels
plt.xlabel("Number of Unique Drug Names per Condition")
plt.ylabel("Number of Conditions")
plt.title("Histogram of Number of Unique Drug Names per Condition")
plt.savefig('unique_drug_names_per_condition-2.png', dpi=300)

# Show the plot
plt.show()
```
