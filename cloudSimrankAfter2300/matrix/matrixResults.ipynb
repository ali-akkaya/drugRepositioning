```python
coExistMatrix = pd.read_csv('searchMatrixCoExistScores.csv', index_col=0)
```

```python
counter = 0
coExistKeysWithData=[]
for key in coExistMatrix.keys():
    if(len(coExistMatrix[key].unique()) > 3):
        print(key)
        coExistKeysWithData.insert(counter, key)
        counter += 1
print(counter)
```

```python
for key in coExistMatrix.keys():
    if(len(coExistMatrix[key].unique()) > 3):
        print(key)
        print(coExistMatrix[key].unique())
```

```python
coExistMatrixReduced = coExistMatrix[coExistKeysWithData]
coExistMatrixReduced.shape
```

```python
coExistMatrixReducedWithReducedIndices = coExistMatrixReduced[coExistMatrixReduced.index.isin(coExistKeysWithData)]
```

```python
coExistMatrixReducedWithReducedIndices.shape
```

```python
coExistMatrixReducedWithReducedIndices.to_csv('coExistMatrixReducedWithReducedIndices.csv')
```
